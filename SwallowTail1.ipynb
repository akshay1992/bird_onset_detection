The following code block contains some helper functions that I will be using later on. Please ignore this if you aren't interested in the inner workings of each function. 

```python
# Helper functions 
def compute_specgram(a, ss=441, ws=2048):    
    # Computes the spectrogram of audio. 
    # (I use this because it less opaque than the built-in function used earlier)
    
    valid = len(a.audio) - ws
    nw = valid//ss
    
    out = np.ndarray((nw,ws),dtype = np.float32)
    window_fn = np.blackman(ws)
    
    for i in xrange(nw):
        # "slide" the window along the samples
        out_slice = np.zeros(ws)
        start = i * ss
        stop = start + ws
        out_slice = np.fft.fft( a.audio[start : stop] * window_fn )
        out[i] = abs(out_slice)
        
    return out

# Get's the index of the frequency bin given frequency in Hz
def getfIndex(a, frequency):
    # a is an audioFile instance 
    if frequency > a.Fs/2:
        print "Error"
    else:
        return frequency * ws / a.Fs
```

The following outlines the procedure used:
* Filter the signal to contain the relavant frequencies where the onset is discernable. This was done in `Audacity` (for now)
* Compute the spectrogram of the filtered signal. 
* Calculate the normalized spectral energy. 
* Fix a `zero-level` in the middle of the dynamic range. (To get both negative and positive peaks)
* Full-wave-rectify the signal about the `zero-level`
* Peak pick the relative minima(here, with an order of `50` - this was hand picked)

```python
# Spectrogram
hop_size = 441
window_size = 2048
spectrogram = compute_specgram(a, ss = hop_size, ws=window_size)

# Spectral energy
energy = spectrogram.sum(axis=1)

plot(energy)
show()
```

```python
# Smooth the spectral energy
kernel = np.blackman(35)       
energy = np.convolve(energy, kernel, mode='same')  

# Normalize the energy signal
norm_energy = energy*1.0/np.max(energy)

plot(norm_energy)
show()

# Notes-to-self 
#  - np.convolve mode is important
#  - kernel 'order' is a custom Input
```

```python
# Fix a zero level roughly in the middle
alpha = 1.2
# zero_level = np.average(norm_energy)*alpha
zero_level = np.sqrt( np.mean( np.square(norm_energy)) ) * alpha

# Note-to-self: this entire block is a custom input
```

```python
# Full-wave-rectify
norm_energy = abs(norm_energy - zero_level)
plot(norm_energy)
show()
```

The minima in this signal is what we want. Each of those distinct minima that you see is the beginning of an onset. 

```python
# Peak picking
peak_pick_order = 50 #Custom Input
onsets_i = argrelmin(norm_energy, order=peak_pick_order)[0]
```

```python
# Plot the results
plot(norm_energy, color='red')

plot_onsets = np.zeros(len(norm_energy))
for element in onsets_i:
    plot_onsets[element] = 0.5
stem(plot_onsets)

show()
```

Clearly beginning and end don't really qualify as onsets. Discarding these. 

```python
start_index = 500 
end_index = 2000

actual_onsets = []
for onset in onsets_i:
    if(onset > start_index and onset < end_index):
        actual_onsets.append(onset)
actual_onsets = np.array(actual_onsets)
```

```python
# Plot the results
plot(norm_energy, color='red')

plot_onsets = np.zeros(len(norm_energy))
for element in actual_onsets:
    plot_onsets[element] = 0.5
stem(plot_onsets)

show()
```

```python
# Convert to seconds
onsets_t = (actual_onsets*hop_size + window_size)*1.0/a.Fs
print onsets_t
```

```python
# Write to file
saveFile = 'onsets_' + a.fileName.strip('_2kto4k.wav') + '.txt'
np.savetxt(saveFile, onsets_t)
```
